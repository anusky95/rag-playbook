# ⚡ Advanced RAG — What's Next After the Basics

## You Know Basic RAG. Now What?

Basic RAG:
```
Question → Search → Get docs → Prompt LLM → Answer
```

This works great. But it fails on:
- Questions that need **multiple search steps** ("What did Alice say about what Bob wrote?")
- Questions where the **LLM should decide** what to search for
- Cases where retrieved docs are **not good enough** and you should search again
- Very **large document sets** where flat search doesn't scale

## Advanced RAG Techniques

| Technique | Core Idea |
|---|---|
| **Agentic RAG** | LLM decides WHAT to search and WHEN to stop |
| **Self-RAG** | LLM checks if retrieved docs are good enough |
| **Multi-Hop RAG** | Chain multiple searches to answer complex questions |
| **RAPTOR** | Summarize chunks into a tree — search at multiple levels |

````python
# Where basic RAG breaks down — a quick demo

# Simple knowledge base
docs = [
    "Alice is the project manager for Project Phoenix.",
    "Project Phoenix is a customer data platform launched in 2023.",
    "Bob is the lead engineer on Project Phoenix.",
    "Bob wrote a design doc saying the database should use PostgreSQL.",
    "Alice approved Bob's database choice in the Q3 review.",
]

# Multi-hop question — needs TWO searches to answer
question = "What did the project manager say about the database?"

# Basic RAG would search for 'project manager' + 'database' in one shot
# That single query might miss that Alice = project manager
# and that Alice approved Bob's database choice

print("Question:", question)
print()
print("To answer this, you need to:")
print("  1. Find who the project manager is → Alice")
print("  2. Find what Alice said about the database → she approved PostgreSQL")
print()
print("Basic RAG tries to do this in ONE search — it often fails.")
print("Agentic / Multi-hop RAG does it in TWO searches — much better.")
````
