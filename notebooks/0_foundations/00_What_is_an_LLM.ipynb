# 🤖 What is an LLM?

## In Plain English

An **LLM (Large Language Model)** is a program that reads text and predicts what comes next.

That's it. Everything else — answering questions, writing code, summarizing — comes from doing that one thing very well on a LOT of text.

```
Input:  "The capital of France is"
Output: "Paris"

Input:  "Write a Python function to add two numbers"
Output: "def add(a, b): return a + b"
```

## How Was It Trained?

```
Step 1: Collect billions of text documents (web, books, code...)
Step 2: Train a neural network to predict the next word
Step 3: Fine-tune it to follow instructions
Step 4: Deploy — now it can answer almost any question!
```

## The Big Limitation

LLMs only know what was in their training data.

```
"What happened in the news today?"  →  ❌ LLM doesn't know
"What's in our company database?"   →  ❌ LLM doesn't know
"Who wrote this internal document?" →  ❌ LLM doesn't know
```

This is exactly the problem **RAG solves** — by giving the LLM access to external knowledge at query time.

````python
# Let's see this in action — a tiny "language model" that predicts the next word
# (This is a toy example, not a real neural network)

# We "train" by counting which words follow other words
training_text = [
    "the cat sat on the mat",
    "the cat ate the fish",
    "the dog sat on the floor",
    "the dog ate the bone",
]

# Build a simple lookup: word → list of words that follow it
next_word_map = {}

for sentence in training_text:
    words = sentence.split()
    for i in range(len(words) - 1):
        current = words[i]
        next_w  = words[i + 1]
        if current not in next_word_map:
            next_word_map[current] = []
        next_word_map[current].append(next_w)

print("Learned next-word predictions:")
for word, next_words in next_word_map.items():
    print(f"  '{word}' → {next_words}")
````

````python
from collections import Counter

# Predict: given a word, what's the most likely next word?
def predict_next(word):
    if word not in next_word_map:
        return "(unknown)"
    # Pick the most common next word
    counts = Counter(next_word_map[word])
    return counts.most_common(1)[0][0]

# Generate a sentence word by word
def generate(start_word, length=5):
    words = [start_word]
    for _ in range(length - 1):
        next_w = predict_next(words[-1])
        words.append(next_w)
    return " ".join(words)

print("Generated sentences:")
print(" ", generate("the"))
print(" ", generate("cat"))
print(" ", generate("dog"))

print("\n💡 Real LLMs do this same thing — just with billions of parameters")
print("   and much more sophisticated pattern matching.")
````

````python
# The knowledge limitation:
# Our toy model only knows what was in the training text

unknown_word = "elephant"
result = predict_next(unknown_word)

print(f"What comes after '{unknown_word}'? → '{result}'")
print()
print("It doesn't know — 'elephant' wasn't in training data.")
print()
print("Real LLMs have the same problem with:")
print("  - Recent events (after training cutoff)")
print("  - Private/internal documents")
print("  - Your specific data")
print()
print("→ Solution: RAG gives the LLM the missing information at query time!")
````
