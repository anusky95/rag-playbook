# 🎯 Tokenization Overview - The Foundation of NLP & RAG

## What is Tokenization?

**Tokenization is the process of breaking text into smaller units (tokens) that models can understand.**

Think of it like this:
- Text: "Hello, World!"
- Tokens: ["Hello", ",", "World", "!"]
- Numbers: [15496, 11, 2159, 0] (token IDs)

## Why Does Tokenization Matter for RAG?

1. **Models don't understand text** - They need numbers
2. **Retrieval efficiency** - Better tokens = better search
3. **Context window limits** - Tokens determine how much text fits
4. **Embedding quality** - Tokenization affects vector representations

## Real-World Applications of Tokenization

- Search engines
- Chatbots
- Machine translation
- Sentiment analysis

## Types of Tokenization (From Simple to Advanced)

### 1. **Word-Level Tokenization** 🔤
- **What**: Split by spaces/punctuation
- **Example**: "Hello world!" → ["Hello", "world", "!"]
- **Pros**: Simple, human-readable
- **Cons**: Huge vocabulary, can't handle unknown words
- **Use Case**: Basic text processing, simple search

### 2. **Character-Level Tokenization** 🔡
- **What**: Each character is a token
- **Example**: "Hi!" → ["H", "i", "!"]
- **Pros**: Tiny vocabulary, handles any text
- **Cons**: Too many tokens, loses meaning
- **Use Case**: Spell correction, low-resource languages

### 3. **Subword Tokenization** 🧩 (MOST IMPORTANT!)

This is what modern AI uses! It's the sweet spot between word and character.

#### **BPE (Byte Pair Encoding)** - GPT's Choice
- **What**: Merges frequent character pairs iteratively
- **Example**: "unhappiness" → ["un", "happiness"] or ["un", "happy", "ness"]
- **Used by**: GPT-2, GPT-3, GPT-4, RoBERTa

#### **WordPiece** - BERT's Choice
- **What**: Similar to BPE but uses likelihood
- **Example**: "unbelievable" → ["un", "##believable"]
- **Used by**: BERT, DistilBERT, E5

#### **SentencePiece** - Universal Approach
- **What**: Language-agnostic, no pre-tokenization needed
- **Example**: Works on raw text including Asian languages
- **Used by**: T5, ALBERT, XLNet, mT5

### 4. **Sentence Tokenization** 📝
- **What**: Split text into sentences
- **Example**: "Hi! How are you?" → ["Hi!", "How are you?"]
- **Use Case**: Document chunking for RAG

## Quick Comparison Table

| Method | Vocabulary Size | Handles Unknown Words | Used By | Best For |
|--------|----------------|----------------------|---------|----------|
| Word-level | 50k-1M+ | ❌ No | Old systems | Simple tasks |
| Character | 50-256 | ✅ Yes | Specialized | Spell check |
| BPE | 32k-50k | ✅ Yes | GPT family | General NLP |
| WordPiece | 30k-50k | ✅ Yes | BERT family | Understanding |
| SentencePiece | 32k-64k | ✅ Yes | T5, mT5 | Multilingual |

````python
# Quick demo of different tokenization approaches
text = "The unhappiness in preprocessing is real!"

# 1. Word-level (simple split)
word_tokens = text.split()
print("Word tokens:", word_tokens)
print("Count:", len(word_tokens))
print()

# 2. Character-level
char_tokens = list(text)
print("Character tokens:", char_tokens)
print("Count:", len(char_tokens))
print()

# 3. Subword (we'll use a simple example)
# In real life, this is learned from data
print("Subword concept:")
print("'unhappiness' could become: ['un', 'happiness'] or ['un', 'happy', 'ness']")
print("This balances vocabulary size with handling unknown words!")
````
