# 🎓 WordPiece Tokenization - BERT's Secret Weapon

## What is WordPiece?

**WordPiece is a subword tokenization algorithm developed by Google, used in BERT and its variants.**

### The Difference from BPE 🤔

```
BPE:       Merges most FREQUENT pairs
WordPiece: Merges pairs that maximize LIKELIHOOD
```

### Why WordPiece? 🎯

- **Better for understanding**: Optimizes for language modeling
- **Used by BERT**: BERT, DistilBERT, RoBERTa, E5, all-MiniLM
- **Great for embeddings**: Perfect for RAG retrieval
- **The ## notation**: Subword pieces are marked with ##

### Example:
```python
"unhappiness" → ["un", "##happiness"]
"playing"     → ["play", "##ing"]
```

The `##` means "this piece is part of the previous word"

## How WordPiece Works

### Algorithm Overview:

1. **Start with characters**: Just like BPE
2. **Calculate likelihood**: For each possible merge
3. **Choose best merge**: Maximize probability of the corpus
4. **Repeat**: Until vocabulary size is reached

### Key Formula:
```
score(x,y) = freq(xy) / (freq(x) × freq(y))
```

Choose the merge with highest score!

## 1. Understanding WordPiece with BERT

````python
# Install transformers if needed
# !pip install transformers

from transformers import BertTokenizer

# Load BERT tokenizer (WordPiece)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

print(f"Vocabulary size: {len(tokenizer)}")
print(f"\nSpecial tokens: {tokenizer.special_tokens_map}")
print(f"\nModel max length: {tokenizer.model_max_length}")
````

````python
# Test WordPiece tokenization
text = "I'm learning about WordPiece tokenization for RAG systems!"

# Tokenize
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text, add_special_tokens=True)

print(f"Text: {text}")
print(f"\nTokens: {tokens}")
print(f"Token IDs: {token_ids}")

# Decode
decoded = tokenizer.decode(token_ids)
print(f"\nDecoded: {decoded}")

# Notice the ## notation!
print("\n💡 Notice: '##ation' means it's part of 'tokenization'")
````

## 2. The ## Notation Explained

````python
# Understanding the ## prefix
examples = [
    "unhappiness",
    "preprocessing",
    "subword",
    "tokenization",
    "embeddingspace"
]

print("WordPiece Tokenization with ## notation:\n")
print(f"{'Word':<20} {'Tokens'}")
print("="*60)

for word in examples:
    tokens = tokenizer.tokenize(word)
    print(f"{word:<20} {tokens}")

print("\n🔑 Rule: Tokens starting with ## are continuations of the previous token")
````

## 3. WordPiece vs BPE Comparison

````python
from transformers import GPT2Tokenizer

# Load both tokenizers
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
gpt2_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Test sentences
test_sentences = [
    "unhappiness",
    "preprocessing",
    "The quick brown fox",
    "I can't believe it's working!"
]

print("WordPiece (BERT) vs BPE (GPT-2):\n")
print(f"{'Text':<30} {'WordPiece':<35} {'BPE'}")
print("="*90)

for text in test_sentences:
    wp_tokens = bert_tokenizer.tokenize(text.lower())
    bpe_tokens = gpt2_tokenizer.tokenize(text)
    print(f"{text:<30} {str(wp_tokens):<35} {str(bpe_tokens)}")

print("\n💡 Notice: WordPiece uses ##, BPE uses Ġ (space marker)")
````

## 4. Special Tokens in WordPiece

````python
# BERT uses special tokens for different tasks
print("Special Tokens in BERT:")
print(f"[CLS]: {tokenizer.cls_token} (ID: {tokenizer.cls_token_id}) - Start of sequence")
print(f"[SEP]: {tokenizer.sep_token} (ID: {tokenizer.sep_token_id}) - Separator")
print(f"[PAD]: {tokenizer.pad_token} (ID: {tokenizer.pad_token_id}) - Padding")
print(f"[UNK]: {tokenizer.unk_token} (ID: {tokenizer.unk_token_id}) - Unknown")
print(f"[MASK]: {tokenizer.mask_token} (ID: {tokenizer.mask_token_id}) - Masked token")

# Example with special tokens
text = "Natural language processing is amazing!"
encoded = tokenizer.encode(text, add_special_tokens=True)
tokens_with_special = tokenizer.convert_ids_to_tokens(encoded)

print(f"\nText: {text}")
print(f"Tokens with special: {tokens_with_special}")
print(f"\n💡 [CLS] at start, [SEP] at end")
````

## 5. Sentence Pair Tokenization (for QA and Retrieval)

````python
# BERT can handle pairs of sentences
question = "What is RAG?"
context = "RAG stands for Retrieval-Augmented Generation."

# Encode sentence pair
encoded = tokenizer.encode_plus(
    question,
    context,
    add_special_tokens=True,
    return_attention_mask=True,
    return_token_type_ids=True
)

tokens = tokenizer.convert_ids_to_tokens(encoded['input_ids'])
token_type_ids = encoded['token_type_ids']

print("Sentence Pair Tokenization:")
print(f"Question: {question}")
print(f"Context: {context}")
print(f"\nTokens: {tokens}")
print(f"Token type IDs: {token_type_ids}")
print("\n💡 Token type: 0=question, 1=context")
````

## 6. Attention Masks and Padding

````python
# Tokenize with padding
texts = [
    "Short text",
    "This is a much longer text that needs more tokens",
]

# Tokenize and pad to same length
encoded = tokenizer(
    texts,
    padding=True,
    truncation=True,
    max_length=20,
    return_attention_mask=True,
    return_tensors='pt'
)

print("Batch Tokenization with Padding:\n")
for i, text in enumerate(texts):
    tokens = tokenizer.convert_ids_to_tokens(encoded['input_ids'][i])
    attention = encoded['attention_mask'][i].tolist()
    
    print(f"Text {i+1}: {text}")
    print(f"Tokens: {tokens}")
    print(f"Attention mask: {attention}")
    print()

print("💡 Attention mask: 1=real token, 0=padding")
````

## 7. WordPiece for RAG: Document Encoding

````python
# Practical RAG example
documents = [
    "Retrieval-Augmented Generation combines retrieval with language models.",
    "The retrieval component searches a knowledge base for relevant information.",
    "The generation component synthesizes the retrieved information into answers.",
    "RAG systems improve accuracy by grounding responses in real data."
]

print("Document Tokenization for RAG:\n")
print(f"{'Doc':<5} {'Tokens':<8} {'First 10 tokens'}")
print("="*70)

all_token_counts = []
for i, doc in enumerate(documents, 1):
    tokens = tokenizer.tokenize(doc)
    token_count = len(tokens)
    all_token_counts.append(token_count)
    
    print(f"{i:<5} {token_count:<8} {tokens[:10]}")

print(f"\nTotal tokens: {sum(all_token_counts)}")
print(f"Average tokens/doc: {sum(all_token_counts)/len(documents):.1f}")
print(f"Max tokens: {max(all_token_counts)}")
````

## 8. Chunking Long Documents

````python
# Long document that exceeds BERT's 512 token limit
long_document = """Retrieval-Augmented Generation is a paradigm that enhances large language 
models by incorporating external knowledge retrieval. This approach addresses the limitation 
of LLMs being constrained to their training data. RAG systems work by first retrieving 
relevant documents from a knowledge base, then using these documents as context for the 
language model to generate responses. The retrieval component typically uses dense embeddings 
and vector similarity search. Popular embedding models include BERT, Sentence-BERT, and E5. 
The generation component can be any LLM like GPT, T5, or BART. RAG systems have shown 
significant improvements in factual accuracy and reduced hallucination compared to 
vanilla language models."""

# Tokenize
tokens = tokenizer.tokenize(long_document)
print(f"Document length: {len(long_document)} characters")
print(f"Token count: {len(tokens)} tokens")
print(f"Exceeds BERT limit (512): {len(tokens) > 512}")

# Chunk the document
def chunk_tokens(tokens, chunk_size=450, overlap=50):
    """Chunk tokens with overlap for context preservation"""
    chunks = []
    start = 0
    
    while start < len(tokens):
        end = start + chunk_size
        chunks.append(tokens[start:end])
        start = end - overlap  # Overlap for context
        
        if end >= len(tokens):
            break
    
    return chunks

chunks = chunk_tokens(tokens, chunk_size=50, overlap=10)
print(f"\nNumber of chunks: {len(chunks)}")
print(f"\nChunk 1 ({len(chunks[0])} tokens): {chunks[0][:15]}...")
print(f"Chunk 2 ({len(chunks[1])} tokens): {chunks[1][:15]}...")
````

## 9. Training Custom WordPiece Tokenizer

````python
# Install tokenizers
# !pip install tokenizers

from tokenizers import Tokenizer
from tokenizers.models import WordPiece
from tokenizers.trainers import WordPieceTrainer
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.processors import TemplateProcessing

# Initialize WordPiece tokenizer
tokenizer_custom = Tokenizer(WordPiece(unk_token="[UNK]"))
tokenizer_custom.pre_tokenizer = Whitespace()

# Define trainer
trainer = WordPieceTrainer(
    vocab_size=1000,
    special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"]
)

# Training corpus (domain-specific)
tech_corpus = [
    "Machine learning models require large datasets for training.",
    "Deep learning uses neural networks with multiple layers.",
    "Natural language processing enables computers to understand text.",
    "Retrieval systems find relevant documents from databases.",
    "Embedding models convert text into dense vector representations.",
    "Attention mechanisms help models focus on important information.",
    "Transformers revolutionized natural language understanding.",
    "Fine-tuning adapts pre-trained models to specific tasks."
] * 50  # Repeat for better learning

# Train
tokenizer_custom.train_from_iterator(tech_corpus, trainer=trainer)

# Add special token processing
tokenizer_custom.post_processor = TemplateProcessing(
    single="[CLS] $A [SEP]",
    pair="[CLS] $A [SEP] $B:1 [SEP]:1",
    special_tokens=[
        ("[CLS]", 1),
        ("[SEP]", 2),
    ],
)

print("✅ Custom WordPiece tokenizer trained!")
print(f"Vocabulary size: {tokenizer_custom.get_vocab_size()}")
````

````python
# Test custom tokenizer
test_text = "Deep learning transforms natural language processing"
output = tokenizer_custom.encode(test_text)

print(f"Text: {test_text}")
print(f"Tokens: {output.tokens}")
print(f"IDs: {output.ids}")
print("\n💡 Notice: Domain-specific terms are efficiently tokenized!")
````

## 10. WordPiece for Embedding Models

````python
# Popular embedding models using WordPiece
from transformers import AutoTokenizer

embedding_models = [
    ('bert-base-uncased', 'BERT Base'),
    ('sentence-transformers/all-MiniLM-L6-v2', 'MiniLM'),
    ('intfloat/e5-base-v2', 'E5 Base'),
]

test_query = "What is retrieval augmented generation?"

print("WordPiece in Popular Embedding Models:\n")
print(f"Query: {test_query}\n")
print(f"{'Model':<40} {'Token Count':<15} {'Tokens'}")
print("="*100)

for model_name, display_name in embedding_models:
    try:
        tok = AutoTokenizer.from_pretrained(model_name)
        tokens = tok.tokenize(test_query)
        print(f"{display_name:<40} {len(tokens):<15} {tokens[:8]}...")
    except:
        print(f"{display_name:<40} {'Not available':<15}")

print("\n🎯 All use WordPiece for consistent tokenization!")
````

## Key Takeaways 🎯

### ✅ WordPiece Advantages:

1. **Likelihood-Based**: Optimizes for language modeling probability
2. **BERT Family**: Used by most popular embedding models
3. **Clean Notation**: ## prefix makes subwords clear
4. **Special Tokens**: Rich set of tokens for different tasks
5. **Sentence Pairs**: Built-in support for QA and similarity

### 🔑 Key Features:

```python
# The ## notation
"playing" → ["play", "##ing"]

# Special tokens
[CLS] ... [SEP] ... [SEP]  # For sentence pairs
[MASK]                      # For masked language modeling

# Token type IDs
0 = first sentence
1 = second sentence
```

### 🚀 For RAG Systems:

**Use WordPiece when:**
- Using BERT-based embedding models
- Need semantic search/similarity
- Building QA systems
- Want interpretable subwords

**Popular models:**
- BERT, DistilBERT, RoBERTa
- Sentence-BERT (all-MiniLM, all-mpnet)
- E5 embeddings
- ELECTRA

### 📊 WordPiece vs BPE:

| Feature | WordPiece | BPE |
|---------|-----------|-----|
| **Algorithm** | Likelihood-based | Frequency-based |
| **Notation** | ## for subwords | Ġ for spaces |
| **Used by** | BERT family | GPT family |
| **Best for** | Understanding | Generation |
| **Typical vocab** | 30k | 50k |
