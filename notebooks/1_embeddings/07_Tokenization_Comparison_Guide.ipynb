# 🎯 Complete Tokenization Comparison Guide - Choose the Right One!

## Quick Decision Tree 🌳

```
What's your use case?
│
├─ Using GPT models? → BPE (GPT-2 tokenizer)
│
├─ Using BERT/embeddings? → WordPiece (BERT tokenizer)
│
├─ Multilingual RAG? → SentencePiece (T5, mT5)
│
├─ Document chunking? → Sentence tokenization (NLTK)
│
├─ Spell checking? → Character-level
│
└─ Simple search? → Word-level
```

## All Methods Side-by-Side

````python
# Install required libraries
# !pip install transformers sentencepiece nltk spacy tiktoken

from transformers import (
    GPT2Tokenizer, 
    BertTokenizer, 
    T5Tokenizer,
    AutoTokenizer
)
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk

# Download NLTK data
nltk.download('punkt', quiet=True)

print("✅ All tokenizers loaded!")
````

## 1. The Ultimate Comparison Test

````python
# Test text with various challenges
test_text = "I can't believe GPT-4's tokenization handles words like 'unhappiness' and 'preprocessing' so well! Dr. Smith's A.I. research at M.I.T. costs $99.99."

print(f"Test Text: {test_text}\n")
print("="*100)

# Method 1: Word-level (NLTK)
word_tokens = word_tokenize(test_text)
print(f"\n1. WORD-LEVEL (NLTK):")
print(f"   Tokens ({len(word_tokens)}): {word_tokens}")

# Method 2: Character-level
char_tokens = list(test_text)
print(f"\n2. CHARACTER-LEVEL:")
print(f"   Tokens ({len(char_tokens)}): {char_tokens[:30]}... (truncated)")

# Method 3: BPE (GPT-2)
gpt2_tok = GPT2Tokenizer.from_pretrained('gpt2')
bpe_tokens = gpt2_tok.tokenize(test_text)
print(f"\n3. BPE (GPT-2):")
print(f"   Tokens ({len(bpe_tokens)}): {bpe_tokens}")

# Method 4: WordPiece (BERT)
bert_tok = BertTokenizer.from_pretrained('bert-base-uncased')
wp_tokens = bert_tok.tokenize(test_text.lower())
print(f"\n4. WORDPIECE (BERT):")
print(f"   Tokens ({len(wp_tokens)}): {wp_tokens}")

# Method 5: SentencePiece (T5)
t5_tok = T5Tokenizer.from_pretrained('t5-small')
sp_tokens = t5_tok.tokenize(test_text)
print(f"\n5. SENTENCEPIECE (T5):")
print(f"   Tokens ({len(sp_tokens)}): {sp_tokens}")

# Method 6: Sentence-level
sent_tokens = sent_tokenize(test_text)
print(f"\n6. SENTENCE-LEVEL (NLTK):")
print(f"   Sentences ({len(sent_tokens)}): {sent_tokens}")
````

## 2. Performance Metrics Comparison

````python
# Comprehensive comparison
import time

test_corpus = "Natural language processing with deep learning is revolutionizing AI. " * 100

tokenizers = {
    "Word (NLTK)": lambda t: word_tokenize(t),
    "Character": lambda t: list(t),
    "BPE (GPT-2)": lambda t: gpt2_tok.tokenize(t),
    "WordPiece (BERT)": lambda t: bert_tok.tokenize(t.lower()),
    "SentencePiece (T5)": lambda t: t5_tok.tokenize(t),
    "Sentence (NLTK)": lambda t: sent_tokenize(t),
}

print("Performance Comparison:\n")
print(f"Corpus: {len(test_corpus)} characters\n")
print(f"{'Method':<25} {'Tokens':<10} {'Vocab Est.':<15} {'Speed'}")
print("="*70)

for name, tokenizer in tokenizers.items():
    # Tokenize
    start = time.time()
    tokens = tokenizer(test_corpus)
    elapsed = time.time() - start
    
    # Estimate vocab size (for this corpus)
    unique_tokens = len(set(tokens)) if tokens else 0
    
    # Known vocab sizes for models
    vocab_sizes = {
        "Word (NLTK)": "50k-1M+",
        "Character": "100-300",
        "BPE (GPT-2)": "50k",
        "WordPiece (BERT)": "30k",
        "SentencePiece (T5)": "32k",
        "Sentence (NLTK)": "N/A",
    }
    
    print(f"{name:<25} {len(tokens):<10} {vocab_sizes[name]:<15} {elapsed*1000:.2f}ms")
````

## 3. RAG Use Case Specific Comparison

````python
# RAG-specific evaluation
rag_document = """Retrieval-Augmented Generation combines retrieval systems with 
large language models. It improves factual accuracy and reduces hallucination."""

rag_query = "How does RAG improve language models?"

print("RAG-Specific Analysis:\n")
print(f"Document: {rag_document}")
print(f"Query: {rag_query}\n")

# Embedding models and their tokenizers
embedding_models = [
    ("BERT Base", "bert-base-uncased", "WordPiece"),
    ("MiniLM-L6", "sentence-transformers/all-MiniLM-L6-v2", "WordPiece"),
    ("E5-base", "intfloat/e5-base-v2", "WordPiece"),
]

print(f"{'Embedding Model':<30} {'Tokenizer':<15} {'Doc Tokens':<12} {'Query Tokens'}")
print("="*80)

for name, model_name, tok_type in embedding_models:
    try:
        tok = AutoTokenizer.from_pretrained(model_name)
        doc_tokens = len(tok.encode(rag_document))
        query_tokens = len(tok.encode(rag_query))
        print(f"{name:<30} {tok_type:<15} {doc_tokens:<12} {query_tokens}")
    except:
        print(f"{name:<30} {tok_type:<15} {'Error':<12} {'Error'}")

print("\n💡 Most RAG systems use WordPiece (BERT family) for embeddings!")
````

## 4. Language Coverage Comparison

````python
# Multilingual test
multilingual_texts = [
    ("English", "Hello, how are you?"),
    ("Spanish", "¡Hola! ¿Cómo estás?"),
    ("Chinese", "你好，你好吗？"),
    ("Japanese", "こんにちは、元気ですか？"),
    ("Arabic", "مرحبا، كيف حالك؟"),
]

print("Multilingual Support Comparison:\n")
print(f"{'Language':<12} {'Text':<30} {'BPE':<8} {'WordPiece':<12} {'SentencePiece'}")
print("="*85)

for lang, text in multilingual_texts:
    bpe_count = len(gpt2_tok.tokenize(text))
    wp_count = len(bert_tok.tokenize(text.lower()))
    sp_count = len(t5_tok.tokenize(text))
    
    print(f"{lang:<12} {text:<30} {bpe_count:<8} {wp_count:<12} {sp_count}")

print("\n🌍 SentencePiece typically handles multilingual best!")
````

## 5. Unknown Word Handling

````python
# Test with rare/unknown words
unknown_words = [
    "supercalifragilisticexpialidocious",
    "pneumonoultramicroscopicsilicovolcanoconiosis",
    "antidisestablishmentarianism",
    "COVID-19",
    "blockchain-based",
]

print("Unknown Word Handling:\n")
print(f"{'Word':<50} {'BPE':<8} {'WordPiece':<12} {'SentencePiece'}")
print("="*85)

for word in unknown_words:
    bpe_tokens = gpt2_tok.tokenize(word)
    wp_tokens = bert_tok.tokenize(word.lower())
    sp_tokens = t5_tok.tokenize(word)
    
    print(f"{word:<50} {len(bpe_tokens):<8} {len(wp_tokens):<12} {len(sp_tokens)}")
    print(f"  BPE: {bpe_tokens[:5]}..." if len(bpe_tokens) > 5 else f"  BPE: {bpe_tokens}")
    print(f"  WordPiece: {wp_tokens[:5]}..." if len(wp_tokens) > 5 else f"  WordPiece: {wp_tokens}")
    print(f"  SentencePiece: {sp_tokens[:5]}...\n" if len(sp_tokens) > 5 else f"  SentencePiece: {sp_tokens}\n")

print("✅ Subword tokenizers handle ANY word by breaking into pieces!")
````

## Quick Reference Summary 📋

### Choose Your Tokenizer:

| Your Task | Use This | Example Code |

| **RAG with BERT embeddings** | WordPiece | `BertTokenizer.from_pretrained('bert-base-uncased')` |
| **RAG with GPT** | BPE | `GPT2Tokenizer.from_pretrained('gpt2')` |

| **Simple search** | Word | `from nltk.tokenize import word_tokenize` |
