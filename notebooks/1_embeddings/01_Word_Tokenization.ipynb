# 📚 Word-Level Tokenization - The Foundation

## What is Word Tokenization?

**Word tokenization splits text into individual words.**

Simple example:
```
"Hello, how are you?" → ["Hello", ",", "how", "are", "you", "?"]
```

## Why Learn This?

- 🎯 **Foundation**: Understanding word tokenization helps you grasp advanced methods
- 🔍 **Search**: Still used in basic keyword search systems
- 📊 **Preprocessing**: First step in many NLP pipelines
- ⚡ **Simple & Fast**: Good for prototyping

## The Problem with Word Tokenization

- ❌ **Huge vocabulary** - Every unique word needs an ID
- ❌ **Unknown words** - What about "supercalifragilisticexpialidocious"?
- ❌ **Multiple forms** - "run", "running", "ran" are different tokens
- ❌ **Language specific** - Doesn't work well for Chinese, Japanese, etc.

## Methods We'll Cover

1. **Simple whitespace split** - Most basic
2. **NLTK word tokenizer** - Handles punctuation better
3. **spaCy tokenizer** - Production-ready
4. **Custom patterns** - For special cases

## 1. Simple Whitespace Tokenization

````python
# The most basic approach
text = "Hello world! This is word tokenization."

# Method 1: Split by spaces
tokens_simple = text.split()
print("Simple split:", tokens_simple)
print("Number of tokens:", len(tokens_simple))

# Problem: Punctuation is attached to words!
# "world!" is one token instead of "world" and "!"
````

````python
# Method 2: Better - split by whitespace and punctuation
import re

def simple_tokenize(text):
    # Split by whitespace and keep punctuation separate
    tokens = re.findall(r'\w+|[^\w\s]', text)
    return tokens

tokens_better = simple_tokenize(text)
print("Better split:", tokens_better)
print("Number of tokens:", len(tokens_better))
````

## 2. NLTK Word Tokenization (Recommended)

````python
# Install NLTK if needed
# !pip install nltk
# python3 -m spacy download en_core_web_sm

import nltk
from nltk.tokenize import word_tokenize

# Download required data (run once)
nltk.download('punkt', quiet=True)

# Tokenize
text = "Don't you think word tokenization is cool? I'm learning RAG!"
tokens_nltk = word_tokenize(text)

print("NLTK tokens:", tokens_nltk)
print("Number of tokens:", len(tokens_nltk))

# Notice: It handles contractions well!
# "Don't" → ["Do", "n't"]
# "I'm" → ["I", "'m"]
````

## 3. spaCy Tokenization (Production-Ready)

````python
# Install spaCy
# !pip install spacy
# !python -m spacy download en_core_web_sm

import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Tokenize
text = "Apple's CEO announced iPhone 15 costs $999.99 in 2024!"
doc = nlp(text)

tokens_spacy = [token.text for token in doc]
print("spaCy tokens:", tokens_spacy)
print("Number of tokens:", len(tokens_spacy))

# Bonus: spaCy gives you extra info!
print("\nToken details:")
for token in doc[:5]:
    print(f"{token.text:15} - POS: {token.pos_:10} - Lemma: {token.lemma_}")
````

## 4. Building a Vocabulary (Word to ID Mapping)

````python
# This is how we convert words to numbers for models
corpus = [
    "I love natural language processing",
    "Natural language processing is amazing",
    "I love learning about NLP and RAG"
]

# Step 1: Tokenize all sentences
all_tokens = []
for sentence in corpus:
    tokens = sentence.lower().split()  # Simple split for demo
    all_tokens.extend(tokens)

# Step 2: Build vocabulary (unique tokens)
vocab = sorted(set(all_tokens))
print(f"Vocabulary ({len(vocab)} words): {vocab}")

# Step 3: Create word to ID mapping
word2id = {word: idx for idx, word in enumerate(vocab)}
id2word = {idx: word for word, idx in word2id.items()}

print("\nWord to ID mapping:")
for word, idx in word2id.items():
    print(f"{word:15} → {idx}")
````

````python
# Step 4: Encode a sentence (words → numbers)
def encode(sentence, word2id):
    tokens = sentence.lower().split()
    # Use 0 for unknown words (not in vocabulary)
    return [word2id.get(token, 0) for token in tokens]

# Encode examples
test_sentence = "I love NLP"
encoded = encode(test_sentence, word2id)
print(f"Sentence: {test_sentence}")
print(f"Encoded: {encoded}")

# Step 5: Decode (numbers → words)
def decode(ids, id2word):
    return [id2word.get(idx, "<UNK>") for idx in ids]

decoded = decode(encoded, id2word)
print(f"Decoded: {' '.join(decoded)}")
````

## 5. The Unknown Word Problem

````python
# What happens with words not in vocabulary?
new_sentence = "I love deep learning and transformers"
encoded_new = encode(new_sentence, word2id)
decoded_new = decode(encoded_new, id2word)

print(f"Original: {new_sentence}")
print(f"Encoded: {encoded_new}")
print(f"Decoded: {' '.join(decoded_new)}")
print("\n⚠️ 'deep', 'learning', 'transformers' became <UNK> (unknown)!")
print("This is why modern models use subword tokenization!")
````

## 6. Practical Example: Text Preprocessing for RAG

````python
# Common preprocessing pipeline
import string

def preprocess_for_rag(text):
    """
    Typical preprocessing for RAG systems
    """
    # 1. Lowercase
    text = text.lower()
    
    # 2. Remove punctuation (optional)
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # 3. Tokenize
    tokens = text.split()
    
    # 4. Remove stopwords (optional)
    stopwords = {'the', 'a', 'an', 'in', 'is', 'it', 'to', 'and'}
    tokens = [t for t in tokens if t not in stopwords]
    
    return tokens

# Example
document = "The quick brown fox jumps over the lazy dog in the park."
processed = preprocess_for_rag(document)

print(f"Original: {document}")
print(f"Processed: {processed}")
print(f"Reduced from {len(document.split())} to {len(processed)} tokens")
````

## 7. Comparing Different Tokenizers

````python
# Compare tokenization methods
test_text = "Dr. Smith earned $50,000 in 2024. He's really happy!"

# Method 1: Simple split
simple = test_text.split()

# Method 2: Regex
regex = simple_tokenize(test_text)

# Method 3: NLTK
nltk_tokens = word_tokenize(test_text)

# Method 4: spaCy
spacy_tokens = [token.text for token in nlp(test_text)]

print("Text:", test_text)
print("\nSimple split:", simple)
print("Regex:", regex)
print("NLTK:", nltk_tokens)
print("spaCy:", spacy_tokens)

print("\n🏆 Winner: spaCy - Handles complex cases best!")
````

## 8. Vocabulary Size Analysis

````python
# Real-world vocabulary sizes
sample_texts = [
    "The cat sat on the mat",
    "The dog sat on the log",
    "Cats and dogs are pets",
    "I have a pet cat and a pet dog",
    "My cat loves playing with my dog"
]

# Count unique words
all_words = []
for text in sample_texts:
    all_words.extend(text.lower().split())

unique_words = set(all_words)

print(f"Total words: {len(all_words)}")
print(f"Unique words (vocabulary): {len(unique_words)}")
print(f"Vocabulary: {sorted(unique_words)}")

print("\n💡 For real applications:")
print("- English vocabulary: ~170,000 words")
print("- Common usage: ~20,000-50,000 words")
print("- This is why subword tokenization is better!")
````

## Key Takeaways 🎯

### ✅ Pros of Word Tokenization:
- Simple to understand and implement
- Human-readable tokens
- Fast for small datasets
- Good for keyword search

### ❌ Cons of Word Tokenization:
- Huge vocabulary size (memory intensive)
- Can't handle unknown words (OOV problem)
- Language-specific rules needed
- Doesn't capture morphology (run, running, ran)

### 🚀 When to Use:
- Simple text analysis
- Keyword-based search
- Exploratory data analysis
- Understanding tokenization concepts

### 🎓 For RAG Systems:
**Don't use word tokenization for modern RAG!** Use:
- BPE (GPT models)
- WordPiece (BERT models)
- SentencePiece (T5, multilingual)
